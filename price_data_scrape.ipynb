```python
import pandas as pd, numpy as np
import yfinance as yf

# select desired time period
start = "2019-01-01"
end   = "2022-02-18"
```

```python
df_financial = yf.download("ETH-USD",  # or input "BTC-USD"
                     #period = "1Y",
                      start= start, 
                      end= end, 
                      interval = "1d",
                      progress=True)
```
